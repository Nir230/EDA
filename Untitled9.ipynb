```python
# 1. Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# 2. Load Data
df = pd.read_csv("titanic.csv")

# 3. Initial Exploration
print(df.info())
print(df.describe())
print(df.isnull().sum())
print(df['sex'].value_counts())
print(df['embarked'].value_counts())

# Drop unnecessary index column
df = df.drop(columns=['Unnamed: 0'])
```

```python
# Histograms
df.hist(bins=20, figsize=(15, 10))
plt.suptitle("Histograms of Numeric Features")
plt.tight_layout()
plt.show()

# Boxplots for numeric features
for col in ['age', 'fare']:
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
# Correlation Heatmap
numeric_df = df[['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare']]
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()

# Pairplot
sns.pairplot(numeric_df, hue='survived')
plt.suptitle("Pairwise Plots", y=1.02)
plt.show()

# Survival rate by sex
sns.barplot(x='sex', y='survived', data=df)
plt.title('Survival Rate by Sex')
plt.show()

# Age distribution by survival
sns.histplot(data=df, x='age', hue='survived', kde=True, multiple='stack')
plt.title('Age Distribution by Survival')
plt.show()
```

```python
# Countplot of Embark Town
sns.countplot(data=df, x='embark_town', hue='survived')
plt.title('Survival by Embark Town')
plt.show()

# Boxplot of Fare by Class
sns.boxplot(data=df, x='class', y='fare')
plt.title('Fare by Passenger Class')
plt.show()
```

âœ… Summary of Findings -

Gender impact: Females had a much higher survival rate than males.

Class differences: Passengers in 1st class had better survival outcomes.

Age insight: Very young and middle-aged passengers had higher survival.

Embark Town: Cherbourg passengers had better survival compared to others.

Correlations: Fare and Pclass had notable relationships with survival.
